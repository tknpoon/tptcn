```python
import pymysql, talib, sys, os
import pandas as pd

#################################
class General:
    conn = None
    ####################
    def __init__(self):
        if self.conn is None:
            self.conn = pymysql.connect(host='tptcn.ddns.net', port=53306, db='p_master',
                                   user=os.environ['MYSQL_READUSER'], password=os.environ['MYSQL_READPASSWORD'],
                                   cursorclass=pymysql.cursors.DictCursor)
    ####################
    def __del__(self):
        if self.conn is not None:
            self.conn.close()
            self.conn = None
            
    ####################
    def worker(self,symbol):
        tmpl = """SELECT  `Date`, `Open` , `High` , `Low` , `Close` , `Volume`
            FROM `consolidated_daily` WHERE `symbol`='%s'
            """
        sql = tmpl % (symbol)
        daily = pd.read_sql(sql, self.conn, index_col=['Date'] )

        rsi = talib.RSI( daily['Close'].values, timeperiod=14)
        daily['RSI'] = rsi
        print daily
#################################

g = General()
g.worker('0011.HK')

```

```python
import pymysql, talib, sys,os
import pandas as pd

conn = pymysql.connect(host='tptcn.ddns.net', port=53306, db='p_master',
                             user=os.environ['MYSQL_READUSER'], password=os.environ['MYSQL_READPASSWORD'],
                             cursorclass=pymysql.cursors.DictCursor)
symbol='1686.HK'
sql = "SELECT  `Date`, `Open` , `High` , `Low` , `Close` , `Volume` FROM `consolidated_daily` WHERE `symbol`='%s' "
daily = pd.read_sql(sql %(symbol),  conn,index_col=['Date'] )

conn.close()
rsi = talib.RSI( daily['Close'].values, timeperiod=14)
daily['RSI'] = rsi
stddev = talib.STDDEV( daily['Close'].values, timeperiod=100)
daily['Close_2sd'] = stddev
print daily

```
